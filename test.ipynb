```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class plot:
    
    # boolean flag to determine if the plot is matplotlib or seaborn
    is_seaborn = False

    def __init__(self, index):
        self.add_index(index)
        self.data = []

    def switch_to_seaborn(self, flag):
        self.is_seaborn = flag

    def add_index(self, index):
        self.index = index.copy()

    def add_data(self, data):
        self.data = data.copy()

    def plot_chart(self, type, color='cyan', thickness=0.8):
        # charts included: line, scatter, bar, histogram, box plot, violin plot, bullet, table, sparkline, connected scatter plot, box, pie, doughnut, gauge, waffle
        if type == 'line':
            sns.lineplot(x=self.index, y=self.data, color=color, linewidth=thickness) if self.is_seaborn else plt.plot(self.index, self.data, color=color, linewidth=thickness)
        elif type == 'scatter':
            sns.scatterplot(x=self.index, y=self.data, color=color) if self.is_seaborn else plt.scatter(self.index, self.data, color=color)
        elif type == 'bar':
            sns.barplot(x=self.index, y=self.data, color=color) if self.is_seaborn else plt.bar(self.index, self.data, color=color)
        elif type == 'histogram':
            sns.histplot(self.data, color=color) if self.is_seaborn else plt.hist(self.data, color=color)
        elif type == 'box plot':
            sns.boxplot(self.data, color=color) if self.is_seaborn else plt.boxplot(self.data, color=color)
        elif type == 'violin plot':
            sns.violinplot(self.data, color=color) if self.is_seaborn else plt.violinplot(self.data, color=color)
        elif type == 'bullet':
            sns.bullet(self.data, color=color) if self.is_seaborn else plt.bullet(self.data, color=color)
        elif type == 'table':
            sns.table(self.data, color=color) if self.is_seaborn else plt.table(self.data, color=color)
        elif type == 'sparkline':
            sns.sparkline(self.data, color=color) if self.is_seaborn else plt.sparkline(self.data, color=color)
        elif type == 'connected scatter plot':
            sns.lineplot(x=self.index, y=self.data, sort=False, color=color) if self.is_seaborn else plt.plot(self.index, self.data, color=color)
        elif type == 'box':
            sns.boxplot(self.data, color=color) if self.is_seaborn else plt.boxplot(self.data, color=color)
        elif type == 'pie':
            plt.pie(self.data, labels=self.index)
        elif type == 'doughnut':
            plt.pie(self.data, labels=self.index, wedgeprops=dict(width=0.5))
        elif type == 'gauge':
            plt.pie(self.data, labels=self.index, wedgeprops=dict(width=0.2))
        elif type == 'waffle':
            plt.pie(self.data, labels=self.index, wedgeprops=dict(width=0.1))
        else:
            print('Invalid chart type')
            
            
    
    def set_grid_params(self, rows, cols, figsize_x, figsize_y):
        self.fig, self.ax = plt.subplots(rows, cols, figsize=(figsize_x, figsize_y))
    
    def plot_grid(self, row, col, type, name_x, name_y, plot_title, color='cyan', thickness=0.8):
        self.ax[row, col].set_xlabel(name_x)
        self.ax[row, col].set_ylabel(name_y)
        self.ax[row, col].set_title(plot_title)

        if type == 'line':
            sns.lineplot(x=self.index, y=self.data, color=color, linewidth=thickness, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].plot(self.index, self.data, color=color, linewidth=thickness)
        elif type == 'scatter':
            sns.scatterplot(x=self.index, y=self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].scatter(self.index, self.data, color=color)
        elif type == 'bar':
            sns.barplot(x=self.index, y=self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].bar(self.index, self.data, color=color)
        elif type == 'histogram':
            sns.histplot(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].hist(self.data, color=color)
        elif type == 'box plot':
            sns.boxplot(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].boxplot(self.data, color=color)
        elif type == 'violin plot':
            sns.violinplot(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].violinplot(self.data, color=color)
        elif type == 'bullet':
            sns.bullet(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].bullet(self.data, color=color)
        elif type == 'table':
            sns.table(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].table(self.data, color=color)
        elif type == 'sparkline':
            sns.sparkline(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].sparkline(self.data, color=color)
        elif type == 'connected scatter plot':
            sns.lineplot(x=self.index, y=self.data, sort=False, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].plot(self.index, self.data, color=color)
        elif type == 'box':
            sns.boxplot(self.data, color=color, ax=self.ax[row, col]) if self.is_seaborn else self.ax[row, col].boxplot(self.data, color=color)
        elif type == 'pie':
            self.ax[row, col].pie(self.data, labels=self.index)
        elif type == 'doughnut':
            self.ax[row, col].pie(self.data, labels=self.index, wedgeprops=dict(width=0.5))
        elif type == 'gauge':
            self.ax[row, col].pie(self.data, labels=self.index, wedgeprops=dict(width=0.2))
        elif type == 'waffle':
            self.ax[row, col].pie(self.data, labels=self.index, wedgeprops=dict(width=0.1))
        else:
            print('Invalid chart type')
            
    def set_x_label(self, label):
        plt.xlabel(label)

    def set_y_label(self, label):
        plt.ylabel(label)

    def set_x_tick_labels(self, labels):
        plt.xticks(self.index, labels)
    
    def set_y_tick_labels(self, labels):
        plt.yticks(self.index, labels)
    
    def figure_size(self, width = 20, height = 5):
        plt.figure(figsize=(width, height))
    
    def set_title(self, title):
        plt.title(title)

    def clear_plot(self):
        plt.clf()
        
    def show_plot(self):
        plt.show()
```
